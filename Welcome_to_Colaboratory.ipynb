## Data science

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
#SEABORN TUTORIAL !!
# whenever we get a dataset, make sure to identify the dependent and independent features.
#if there is only 1 feature, we call it univariate analysis and if 2 , bivariate analysis. 
# Joinplot is used for bivariate analysis.

import seaborn as sns
df = sns.load_dataset("tips")
df.head()                         # tip, size is a dependent feature.   
```

```
df.info()
```

```
# Correlation with heat map. SNS has a property called as heatmap which helps us to find out the correlation between features.
#this is used to show a 2d correlation matrix(table) between 2 discrete dimensions  or event types. imp in feature selection
#correlation can only be found out, if the values are int or float,, not categorical as they are object type. and the value
# ranges from -1 to +1,,,pearson correlation.
df.corr()  
```

```
df.dtypes
```

```
sns.heatmap(df.corr())
```

```
# JOIN PLOT: it allows the study the relationship between 2 numeric variable. Central chart display their correlation.
# it is usually hexbin, scatter plot or 2d density function or 2D histogram.

sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```
# PAIRPLOT: this is known as scatter plot.. in which one variable in the same data row is matched with another variable's value.
# it shows all the correlation between different variable of the kind..(either numeric or categorical)
sns.pairplot(df)
```

```
df['smoker'].value_counts()
```

```
# if we need to add another feature which is categorical,, we would use the 'hue' parameter.
sns.pairplot(df, hue = 'sex')
```

```
sns.displot(df['tip'], kde = False, bins = 10)  #kde = kernel density estimation   
```

```
#Categorical PLOTs:  box, violin, count, bar are for categorical plots.
#Count plot. 
sns.countplot('day', data = df)
```

```
#BAr plot
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```
#Box plot: a box and whisker plot is a graph that represents information in 5 number summary.
sns.boxplot(x='day', y = 'total_bill', data = df)
```

```
sns.boxplot(data = df, orient = 'v')
```

```

sns.boxplot(x = "tip", y = "day", hue = "smoker", data = df)
```

```
# Violin  plot : it helps us to see distribution in data in terms of both kernel density and box plot
# its nothing but a combination of both kde and box plot
sns.violinplot(x = 'total_bill', y = 'day', data = df)
```
